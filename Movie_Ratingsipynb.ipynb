```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/tmdb_5000_movies.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df['rating'] = df['vote_average']
```

```python
df.describe()
```

```python
mean_rating = df['vote_average'].mean()
mode_rating = df['vote_average'].mode(0)
median_rating = df['vote_average'].median()

print("mean:",mean_rating)
print("mode:",mode_rating)
print("median:",median_rating)
```

```python
plt.figure()
plt.hist(df['vote_average'], bins=10)
plt.title('Distribution of  Movie rating')
plt.xlabel('vote_average')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
plt.boxplot(df['vote_average'])
plt.title('Box plot of rating')
plt.show()
```

```python
top_movies = df.sort_values(by = 'vote_average', ascending = False)
top_movies = top_movies[['title', 'vote_average']].head(10)
```

```python
import ast

# Convert the string representation of lists to actual lists of dictionaries
df['genres_list'] = df['genres'].apply(ast.literal_eval)

# Extract only the genre names
df['genre_names'] = df['genres_list'].apply(lambda x: [d['name'] for d in x])

# Explode the list of genre names into separate rows
df_exploded_genres = df.explode('genre_names')

# Now group by the 'genre_names'
genre_ratings = df_exploded_genres.groupby('genre_names')['vote_average'].mean().sort_values(ascending=False)

print(genre_ratings.head(10))
```

```python
genre_ratings.head(10).plot(kind='bar')
plt.title("Top Genres by Average Rating")
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure(figsize=(10,5))

# Histogram
plt.subplot(1,2,1)
plt.hist(df['rating'], bins=10)
plt.title("Ratings Distribution")

# Box Plot
plt.subplot(1,2,2)
sns.boxplot(x=df['rating'])
plt.title("Box Plot")

plt.tight_layout()
plt.show()
```
